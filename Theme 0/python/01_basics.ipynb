````python
# üìå –ü—Ä–æ–≥—Ä–∞–º–∞ –¥–ª—è –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ä—ñ–≤–Ω—è –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ –Ω–∞ –æ—Å–Ω–æ–≤—ñ –≤–≤–µ–¥–µ–Ω–æ–≥–æ —á–∏—Å–ª–∞

# –ö—Ä–æ–∫ 1: –ó–∞–ø–∏—Ç—É—î–º–æ —É –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞
try:
    # –§—É–Ω–∫—Ü—ñ—è input() –æ—Ç—Ä–∏–º—É—î —Ç–µ–∫—Å—Ç –≤—ñ–¥ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞, –∞ int() –ø–µ—Ä–µ—Ç–≤–æ—Ä—é—î –π–æ–≥–æ —É —Ü—ñ–ª–µ —á–∏—Å–ª–æ
    losses = int(input("–í–≤–µ–¥—ñ—Ç—å –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞: "))

    # –ö—Ä–æ–∫ 2: –ê–Ω–∞–ª—ñ–∑—É—î–º–æ, –¥–æ —è–∫–æ–≥–æ —Ä—ñ–≤–Ω—è –Ω–∞–ª–µ–∂–∞—Ç—å —Ü—ñ –≤—Ç—Ä–∞—Ç–∏

    # –Ø–∫—â–æ –≤—Ç—Ä–∞—Ç–∏ –º–µ–Ω—à—ñ –∑–∞ 50 ‚Äì —Ä—ñ–≤–µ–Ω—å –≤–≤–∞–∂–∞—î—Ç—å—Å—è –Ω–∏–∑—å–∫–∏–º
    if losses < 50:
        print("üîπ –†—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞: –Ω–∏–∑—å–∫–∏–π.")

    # –Ø–∫—â–æ –≤—Ç—Ä–∞—Ç–∏ –≤—ñ–¥ 50 –¥–æ 149 –≤–∫–ª—é—á–Ω–æ ‚Äì —Å–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å
    elif 50 <= losses < 150:
        print("üî∏ –†—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞: —Å–µ—Ä–µ–¥–Ω—ñ–π.")

    # –Ø–∫—â–æ –≤—Ç—Ä–∞—Ç–∏ –≤—ñ–¥ 150 –¥–æ 299 –≤–∫–ª—é—á–Ω–æ ‚Äì –≤–∏—Å–æ–∫–∏–π —Ä—ñ–≤–µ–Ω—å
    elif 150 <= losses < 300:
        print("‚ö†Ô∏è –†—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞: –≤–∏—Å–æ–∫–∏–π.")

    # –Ø–∫—â–æ –≤—Ç—Ä–∞—Ç–∏ 300 –∞–±–æ –±—ñ–ª—å—à–µ ‚Äì –∫—Ä–∏—Ç–∏—á–Ω–∏–π —Ä—ñ–≤–µ–Ω—å
    else:
        print("üö® –†—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞: –∫—Ä–∏—Ç–∏—á–Ω–∏–π.")

# –ö—Ä–æ–∫ 3: –Ø–∫—â–æ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á –≤–≤—ñ–≤ –Ω–µ —á–∏—Å–ª–æ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, —Ç–µ–∫—Å—Ç), —Ç–æ –±—É–¥–µ –ø–æ–º–∏–ª–∫–∞
except ValueError:
    # –ü–æ–≤—ñ–¥–æ–º–ª—è—î–º–æ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—É –ø—Ä–æ –ø–æ–º–∏–ª–∫—É
    print("‚ùå –ü–æ–º–∏–ª–∫–∞: –≤–≤–µ–¥—ñ—Ç—å, –±—É–¥—å –ª–∞—Å–∫–∞, —Ü—ñ–ª–µ —á–∏—Å–ª–æ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥: 120).")
````
